```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("C:/Users/syaki/Documents/~SIDE WORK/Polynomial-Salary/Position_Salaries.csv")
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
dataset.head(10)
```

```python
# training Linear Regression model
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
poly_reg3 = PolynomialFeatures(degree=3)
X_poly3 = poly_reg3.fit_transform(X)
lin_reg3 = LinearRegression()
lin_reg3.fit(X_poly3,y)
```

```python
poly_reg4 = PolynomialFeatures(degree=4)
X_poly4 = poly_reg4.fit_transform(X)
lin_reg4 = LinearRegression()
lin_reg4.fit(X_poly4,y)
```

```python
# visualize Linear Regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# visualize Polynomial Regression
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# visualize Polynomial Regression (for higher resolution and smoother curve)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,y,color='black')
plt.plot(X,lin_reg3.predict(X_poly3),color='blue')
plt.title("Polynomial Regression degree 3")
plt.xlabel("Positions")
plt.ylabel("Salaries")
plt.show()
```

```python
plt.scatter(X,y,color='black')
plt.plot(X,lin_reg4.predict(X_poly4),color='red')
plt.title("Polynomial Regression degree 4")
plt.xlabel("Positions")
plt.ylabel("Salaries")
plt.show()
```

```python
plt.scatter(X,y,color='black')
plt.plot(X,lin_reg.predict(X),color='aqua')#degree=1
plt.plot(X,lin_reg2.predict(X_poly),color='green') #degree=2
plt.plot(X,lin_reg3.predict(X_poly3),color='blue')#degree=3
plt.plot(X,lin_reg4.predict(X_poly4),color='red')#degree=4

plt.title("Polynomial Regression V/s Linear Regression")
plt.xlabel("Positions")
plt.ylabel("Salaries")
plt.show()
```
